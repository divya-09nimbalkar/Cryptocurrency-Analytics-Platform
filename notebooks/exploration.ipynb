```python
import pandas as pd
import matplotlib.pyplot as plt
```

```python
df = pd.read_csv(r"D:\Projects all\All GITHUB PROJECTS TO UPLOAD\Z FOLDERS OF ALL THE MAIN TOPICS\Data_Engineering\Cryptocurrency_Analytics_Platform\data\raw\sample.csv")
df.head()
```

```python
# Cell 3: Basic summary
print("Rows:", len(df))
print("Columns:", df.columns.tolist())
print("Average Price:", df["price"].mean())
print("Average Volume:", df["volume"].mean())
```

```python
# Cell 4: Plot price trend
plt.figure(figsize=(8,5))
plt.plot(df["date"], df["price"], marker="o")
plt.title("Cryptocurrency Price Trend")
plt.xlabel("Date")
plt.ylabel("Price")
plt.xticks(rotation=45)
plt.show()
```

```python
# Cell 5: Plot volume trend
plt.figure(figsize=(8,5))
plt.bar(df["date"], df["volume"])
plt.title("Cryptocurrency Volume Trend")
plt.xlabel("Date")
plt.ylabel("Volume")
plt.xticks(rotation=45)
plt.show()
```

```python
# Cell 6: Correlation analysis
df_numeric = df.select_dtypes(include=["number"])
df_numeric.corr()
```

```python
df["date"] = pd.to_datetime(df["date"])
df.set_index("date", inplace=True)
df["price"].plot(figsize=(10,6), title="Price Over Time")
plt.show()
```

```python
# Cell 7: Moving average
if "price" in df.columns and "date" in df.columns:
    df["price_ma"] = df["price"].rolling(window=2).mean()
    plt.figure(figsize=(8,5))
    plt.plot(df["date"], df["price"], marker="o", label="Price")
    plt.plot(df["date"], df["price_ma"], marker="o", linestyle="--", label="2-day MA")
    plt.title("Price with Moving Average")
    plt.xlabel("Date")
    plt.ylabel("Price")
    plt.legend()
    plt.xticks(rotation=45)
    plt.show()
```

```python
# Cell 8: Volatility
if "price" in df.columns and "date" in df.columns:
    df["price_change_pct"] = df["price"].pct_change() * 100
    plt.figure(figsize=(8,5))
    plt.bar(df["date"], df["price_change_pct"])
    plt.title("Daily Price Change (%)")
    plt.xlabel("Date")
    plt.ylabel("Change (%)")
    plt.xticks(rotation=45)
    plt.show()
```

```python
# Cell 9: Price vs Volume scatter
if "price" in df.columns and "volume" in df.columns:
    plt.figure(figsize=(6,6))
    plt.scatter(df["price"], df["volume"], c="blue", alpha=0.7)
    plt.title("Price vs Volume")
    plt.xlabel("Price")
    plt.ylabel("Volume")
    plt.show()
```
